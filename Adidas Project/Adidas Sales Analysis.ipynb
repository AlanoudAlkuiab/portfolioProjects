## Import necessary libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk 
```

```python
#read df
df=pd.read_excel("Adidas US Sales Datasets.xlsx")
```

```python
#print the first 5 rows of the DataFrame
df.head()
```

```python
#making the row index 3 as columns
df.columns=df.iloc[3]
```

```python
#drop the first 3 rows 
df = df.drop(index=[0, 1,2,3])
```

```python
#drop the first column
df=df.iloc[:,1:]
```

```python
#reset index
df = df.reset_index(drop=True)
```

```python
#print the first 5 rows of the DataFrame
df.head()
```

```python
#print a summary of the DataFrame
df.info()
```

```python
df.describe()
```

# Data Preprocessing 

### Handling data types

```python
# change the dtypes of some columns
df[['Price per Unit','Total Sales','Operating Profit','Operating Margin']] = df[['Price per Unit','Total Sales','Operating Profit','Operating Margin']].astype('float')
df['Units Sold'] = df['Units Sold'].astype('int')
#date to datetime
df['Invoice Date']=pd.to_datetime(df['Invoice Date'])
```

### Checking for missing values

```python
#checking the number of missing values in df
df.isna().sum()
```

Since there is no missing values in df , i will leave the df as it is . 

### Checking for  duplicate values 

```python
df.duplicated().sum()
```

There is no duplicated data in df 

### Creation of new columns 

```python
#extracting the month and the day and the year in a new columns
df['Year']=df['Invoice Date'].dt.year
df['Month']=df['Invoice Date'].dt.month
df['Day']=df['Invoice Date'].dt.day

```

```python
#checking for the number of Regions in the data
df['Region'].value_counts()
```

```python
#checking for any splling error in the Region columns
df['City'].unique()
```

# EDA : Exploratory Data Analysis

### Question : which retailers generate  high sales for Adidas ?

```python
#grouping the retailers total sales 
R_TotalSales=df.groupby(['Retailer'])['Total Sales'].sum().sort_values(ascending=False).to_frame()
```

```python
#print the result
R_TotalSales
```

```python
# Create a bar plot
sns.barplot(x=R_TotalSales.index, y=R_TotalSales['Total Sales'].values, data=R_TotalSales)

# Customize the plot
sns.set_style('darkgrid')
sns.set_palette('bright')

# Add a title and labels
plt.title('Total Sales Of Adidas By Retailer ')
plt.xlabel('Retailers')
plt.ylabel('Total Sales in millions')

# Customize the appearance of the bars
plt.bar_width = 0.4
plt.bar_edgecolor='white'


# Show the plot
plt.show()
```

##### Answer :
As shown in the graph the retailers that generate high sales are West Gear with over 2.4M\\$ in total sales , Foot locker with over 2.2M\\$ , and the rest of the retailers are generating sales between 1.8M\\$ to 700K\\$ .  

```python
#Percentage of total sales of Adidas in different retailers 
R_TotalSales.plot(kind='pie',subplots=True,figsize=(12,7),
title='Total sales per Retailer',autopct='%1.0f%%')
```

### Question : What is the price disturibution of Adidas products ? 

```python
df['Price per Unit'].plot(kind='hist',bins=25,figsize=(12,7),title='Products price distribution')
```

##### Answer :
the price range of  Adidas products lies between 40\\$ and 60\\$ , with a mean of 45.22\\$


### Question : which product generae the the highest profit for  Adidas compare to the number of the unit sold? 


```python
P_profit=df.groupby(['Product'])['Operating Profit'].sum().sort_values(ascending=False).to_frame()
```

```python
P_profit.plot.bar(figsize=(8, 5))

# Create a bar plot
sns.barplot(x=P_profit.index, y=P_profit['Operating Profit'].values, data=P_profit)

# Customize the plot
sns.set_style('darkgrid')
sns.set_palette('bright')

# Add a title and labels
plt.title('Profit By Product ')
plt.xlabel('Products')
plt.ylabel('Profit in Millions')

# Customize the appearance of the bars
plt.bar_width = 0.4
plt.bar_edgecolor='white'


# Show the plot
plt.show()
```

##### Answer :
 the product that generate the highest profit for Adidas is Men's Street Footwear with over 8M\\$ in profit .

```python
#Percentage of Operating profit of products  
palette_color = sns.color_palette('bright')
  
# plotting data on chart
plt.pie(P_profit['Operating Profit'], labels=P_profit.index, colors=palette_color, autopct='%.0f%%')
  
# displaying chart
plt.show()
```

### َQuestion : wihch regions generate higher profit ?

### Question : profit distribution over time (Month) ?  


```python
#grouping data by Month and Year 
sales_by_month = df.groupby(['Year','Month'])['Total Sales'].sum().reset_index()


# Set a new color palette for hue
palette=sns.color_palette('dark:#5A9',as_cmap=True)

# Plot line chart
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
sns.lineplot(data=sales_by_month, x='Month', y='Total Sales',hue='Year',marker='o',palette=palette)
```

##### Answer : 
As shown in the graph there was decline in the profit in the middle of 2020 till the end of the year 2020 this is maybe due to Covid-19 since the majority of Adidas  sales comes from in-store , and at that time the store were closed , after that the profit start to increase in 2021 with minor decline 


### Question : which city that generates hgihest  average profit ?
